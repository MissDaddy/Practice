```python
# The major Python visualization library is matplotlib.

# Import its sub-libray .pyplot under alias 'plt'.

import matplotlib.pyplot as plt

# Use magic line to display plots inline.

%matplotlib inline
```

```python
# To be able to plot data values contained in dataframes, import the pandas library first.

import pandas as pd

# Let's hand code a small dataset and store it in the votes variable. Note we are using dictionaries as arguments.

votes = pd.DataFrame({'name':['Obama', 'Trump', 'HClinton', 'TCruz'], 'vote':[35,25,20,20]})

votes
```

```python
# We'll use a wrapper plot() function in this case.

votes.plot.pie('vote', labels=votes['name'])

# votes.plot(kind='pie', labels=votes['name']) generates the same results
```

```python
# To increase the size of the pie, set the figsize argument to equal aspect ratio, using a tuple of 2 equal values.

# Add data values to pie slices with the autopct argument set to the percentage listed as a fraction


votes.plot.pie('vote',labels=votes['name'], autopct="%1.1f%%", figsize=(8,8))

# To get rid of the decimal places, set autopct="%1.f%%"

# To rotate the pie, set startangle=90 or any angle you wish, e.g. 45, 53, 120 etc.
```

```python
# You can choose your colors instead of the default ones.

# Define a list of colors.
colors =['lightgrey', 'grey', 'darkgrey', 'black']

# Inside plot() pass the argument colors set to the color list you defined above.

votes.plot.pie('vote',labels=votes['name'], autopct="%1.1f%%", figsize=(8,8), colors=colors)
```

```python
# Let's dissect a pie to improve its legibility or to make one part stand out.
# Set the argument explode to 4 values, one of them should be a float value to explode out one pie slice
# shadow=True will add shadows to pie slices.

votes.plot.pie('vote',labels=votes['name'], autopct="%1.1f%%",figsize=(8,8),explode=(0.15, 0, 0, 0), shadow=True)
```

```python
# Histograms are perfect for one variable, best for continuous data.
# Yet histograms may as well suit discrete datasets whose values are almost continuous.

# Let's hand code a list of wine ages.

wine_age =[20, 45, 40, 42, 5, 1, 2, 23, 100]

wine_age
```

```python
# Let's create a generic function that is meant to plot numeric values as a histogram.
# The statement def is used to create new functions.
# Call that function Age and pass the argument a.
# Pass the argument bins and set it to 5 values, which will split the data into 5 intervals called 'bins'.
# Use the .plt object this time.

def Age(x):   
    plt.hist(x, bins=5)
    plt.ylabel('Age')
    plt.title('Distribution by age')

# To see what the our defined function produces on the specific data, call the function by its name 'Age' and 
# set the argument 'a'  to the variable 'wine_age'.
# The function 'Age' will plot the wine_age list as a histogram

Age(x=wine_age)
```

```python
# Let's hand code one more list that contains the age of employees.

employees_age =[22, 50, 45, 40, 23, 25, 60, 32, 35, 36, 36, 41, 32, 21, 39, 37]

employees_age

# Note that the length of this list is longer than the list 'wine_age'. In other words, it contains more data points.
# Because the employee data is similar to the wine data, we can recylce the Age function and apply it to both data sets  
```

```python
# Let's print out the plot generated by the function Age.

Age(x=employees_age)
```

```python
# Let's read in the xlsx file called 'books.xlsx'. The file should be in your directory.

# in PyClass2, we used the pandas method .read_csv(). But for .xlsx files, use the mehod .read_excel().

books = pd.read_excel('books.xlsx', dtype=object)

books
```

```python
# Remove the empty rows the same way we did it in PyClass6.

booksClean = books.dropna(axis=0)

booksClean
```

```python
# Get the summary of the data with describe().

booksClean.describe()
```

```python
# Use the wrapper plot() to plot all numeric data as bars.

booksClean.plot(kind='bar', figsize=(10, 5))
```

```python
# Write an empty function that is suited for a single purpose of plotting all books by their height.

def books_bar():
    booksClean.plot(kind='bar', title='Books by height', x='Title', legend=False, figsize=(20, 12))
    plt.xlabel('Titles', fontsize=20, color='red')
    plt.ylabel('Height', fontsize=20, color='red')


books_bar()

# Try to set kind to 'barh' or line' or 'area' or 'density'
```

```python
# We can count with len() how many items the 'Genre' column contains.

genre=len(booksClean['Genre'])

genre
```

```python
# Yet we already know that the data categories repeat in that and a few other columns.

# We can count how many unique data categories are in the 'Genre' column with nunique().

booksClean['Genre'].nunique()
```

```python
# Compare nunique() with unique(). The latter will return the categories themselves.

booksClean['Genre'].unique()
```

```python
# We can count the number of values per each unique category in the 'Genre' column with value_counts().

No_books = booksClean['Genre'].value_counts()

No_books
```

```python
# We'll need the numpy method arrange() to plot the data of the 'Genre' column. So let's import it first.

import numpy as np

# Write the function to count values of unique genre categories and plot those values by unique categories.
# Use np.arange(5) to break your data as evenly spaced 5 parts. the arange function belongs to numpy

def genre_bar():
    
    data=np.arange(5) #returns data as 5 evenly spaced parts
    No_books = booksClean['Genre'].value_counts() # count the frequencies (values) per category
    plt.bar(data, No_books, color=('yellow', 'green', '#f44b89', 'blue', '#ff0000')) #plot 5 bars on x axis and freq on y
    plt.xlabel('Genre', fontsize = 14)
    plt.ylabel('No of Books', fontsize=14)
    plt.xticks(data, labels=('fiction', 'nonfiction', 'tech', 'science', 'philosophy'), fontsize=10)
    plt.title('Distribution of Books by Genre', fontsize=17, color='darkblue')

genre_bar()
```
